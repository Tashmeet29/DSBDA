```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('DSBDALExam DataSets/Adult/adult.csv', header=None)
column_names = ['age', 'workclass', 'fnlwgt', 'education', 'education-num', 'marital-status', 'occupation',
                'relationship', 'race', 'sex', 'capital-gain', 'capital-loss', 'hours-per-week', 'native-country', 'loh50']
df.columns = column_names
df.index.name = 'Id'
df.head()
```

```python
plt.figure(figsize=(12,8))
df.hist(bins=20, color='lightblue', edgecolor='black', grid=False)
plt.tight_layout()
plt.show()
```

```python
# Histogram
plt.figure(figsize=(12,8))
df.hist(column='age', bins=20, color='lightblue', edgecolor='black', grid=False)
plt.tight_layout()
plt.show()
```

```python
# Dot Plot
plt.figure(figsize=(12, 8))
sns.stripplot(x='education', data=df, jitter=True)
plt.xlabel('Education Level')
plt.ylabel('Frequency')
plt.title('Dot Plot of Education Level')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 8))
marital_status_counts = df['marital-status'].value_counts()
sns.barplot(x=marital_status_counts.index, y=marital_status_counts.values, palette='pastel')
plt.xlabel('Marital Status')
plt.ylabel('Count')
plt.title('Bar Plot of Marital Status')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 8))
plt.plot(df['age'], label='Age')
plt.plot(df['hours-per-week'], label='Hours per Week', color='lightblue')
plt.xlabel('Id')
plt.ylabel('Value')
plt.title('Line Chart of Age and Hours per Week')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(23, 8))
sns.boxplot(x='age', y='hours-per-week', data=df, color='lightblue', width=0.2)
sns.histplot(df['age'], color='skyblue', bins=20, kde=True)
sns.scatterplot(x=df['age'], y=df['hours-per-week'], color='red', marker='o', alpha=0.5)
plt.xlabel('Age')
plt.ylabel('Hours per Week')
plt.title('Distribution of Age with Scatter Plot and Box Plot')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
income_counts = df['loh50'].value_counts()
sns.barplot(x=income_counts.index, y=income_counts.values, palette='pastel')
plt.xlabel('Income Level')
plt.ylabel('Count')
plt.title('Distribution of Income Level')
plt.tight_layout()
plt.show()
```
