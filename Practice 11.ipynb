```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df.hist(figsize=(10,12))
plt.tight_layout()
plt.show()
```

```python
merged_dataset = pd.merge(df, ratings, left_index=True, right_index=True, how='inner')
merged_dataset
```

```python
desired_columns = ['id', 'original_title', 'original_language', 'movieId', 'rating', 'budget', 'vote_average', 'revenue']
merged_dataset[desired_columns].head()
```

```python
sns.pairplot(merged_dataset[desired_columns], diag_kind='hist')
plt.show()
```

```python
plt.figure(figsize=(16, 10))
df['original_language'].value_counts().plot(kind='bar')
plt.title('Number of Movies by Original Language')
plt.xlabel('Original Language')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='vote_average', y='rating', data=merged_dataset[desired_columns])
plt.title('Scatter Plot of Average Rating vs. Rating')
plt.xlabel('Average Rating')
plt.ylabel('Rating')
plt.show()
```
