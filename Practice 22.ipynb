```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('DSBDALExam DataSets/Adult/adult.csv', header=None)
column_names = ['age', 'workclass', 'fnlwgt', 'education', 'education-num', 'marital-status', 'occupation',
                'relationship', 'race', 'sex', 'capital-gain', 'capital-loss', 'hours-per-week', 'native-country', 'loh50']
df.columns = column_names
df.index.name = 'Id'
df.head()
```

```python
# Histogram
plt.figure(figsize=(12,8))
df.hist(bins=20, color='lightblue', edgecolor='black', grid=False)
plt.tight_layout()
plt.show()
```

```python
# Histogram
plt.figure(figsize=(12,8))
df.hist(column='education-num', bins=20, color='lightblue', edgecolor='black', grid=False)
plt.tight_layout()
plt.show()
```

```python
# Histogram
plt.figure(figsize=(12,8))
df.hist(column='age', bins=20, color='lightblue', edgecolor='black', grid=False)
plt.tight_layout()
plt.show()
```

```python
# Pie Charts
plt.figure(figsize=(8, 8))
race_counts = df['race'].value_counts()
plt.pie(race_counts, labels=race_counts.index, autopct='%1.1f%%', colors=sns.color_palette('pastel'))
plt.title('Distribution of Race')
plt.tight_layout()
plt.show()
```

```python
# Box Plot
plt.figure(figsize=(12, 8))
sns.boxplot(data=df[['age', 'hours-per-week']], palette='pastel')
plt.title('Box Plot of Age and Hours per Week')
plt.ylabel('Values')
plt.tight_layout()
plt.show()
```

```python
# Scatter-plot
plt.figure(figsize=(10,8))
sns.scatterplot(x='age', y='education-num', data=df, alpha=0.6)
plt.xlabel('Age')
plt.ylabel('Education Number')
plt.title('Scatterplot of Age vs Education num')
plt.tight_layout()
plt.show()
```

```python
# Combined plot (Boxplots on Scatterplot) - Adding boxplots of 'age' and 'education-num' to the scatter plot
plt.figure(figsize=(16,8))
sns.scatterplot(x='age', y='education-num', data=df, alpha=0.6)
sns.boxplot(x='age', y='education-num', data=df, color='lightblue', width=0.2)
plt.xlabel('Age')
plt.ylabel('Education Number')
plt.title('Scatter Plot of Age vs Education Number with Boxplots')
plt.tight_layout()
plt.show()
```
