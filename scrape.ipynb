```python
import requests
from bs4 import BeautifulSoup
import time

def scrape_reviews(url):
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'}

    # Introduce a delay to avoid being blocked
    time.sleep(2)

    # Send a request to the URL with headers
    response = requests.get(url, headers=headers)

    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')

        # Extracting reviews
        reviews = []
        review_elements = soup.find_all('div', class_='a-section review aok-relative')

        for review_element in review_elements:
            review = {
                'customer_name': review_element.find('span', class_='a-profile-name').text.strip(),
                'rating': review_element.find('span', class_='a-icon-alt').text.split()[0],
                'review': review_element.find('span', class_='a-size-base review-text review-text-content').text.strip()
            }
            reviews.append(review)

        return reviews
    else:
        print(f"Failed to retrieve data. Status code: {response.status_code}")
        return None

if __name__ == "__main__":
    # Amazon URL for OnePlus product reviews
    amazon_url = 'https://www.amazon.in/Redmi-Startrail-Storage-MediaTek-Dimensity/product-reviews/B0CNX6WVG5/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews'

    # Scrape and print the reviews
    scraped_reviews = scrape_reviews(amazon_url)

    if scraped_reviews:
        print("Number of Reviews:", len(scraped_reviews))
        for idx, review in enumerate(scraped_reviews, start=1):
            print(f"Review #{idx}:")
            print(f"Customer Name: {review['customer_name']}")
            print(f"Rating: {review['rating']}")
            print(f"Review: {review['review']}")
            print("\n" + "="*50 + "\n")
    else:
        print("No reviews were scraped.")
```

```python
import requests
import bs4
```

```python
request1 = requests.get('https://www.flipkart.com/vivo-t2-pro-5g-dune-gold-256-gb/p/itm1230688cdef18?pid=MOBGT4RZUPKD8HDK&lid=LSTMOBGT4RZUPKD8HDK4QMUBF&marketplace=FLIPKART&cmpid=content_mobile_8965229628_gmc')
```

```python
request1
```

```python
soup = bs4.BeautifulSoup(request1.text)
```

```python
reviews = soup.find_all('div', {'class': 'ZmyHeo'})
for review in reviews:
    print(review.get_text() + "\n\n")
```

```python
customer_name = soup.findAll('p', {'class': '_2NsDsF AwS1CA'});
for cust_name in customer_name:
    print(cust_name.get_text() + "\n")
```

```python
soup = bs4.BeautifulSoup(request1.text)
reviews = []
for review_element in soup.find_all('div', class_='RcXBOT'):
    customer_name = review_element.find('p', class_='_2NsDsF AwS1CA')
    rating_elem = review_element.find('div', class_='XQDdHH Ga3i8K')
    review_text = review_element.find('div', class_='ZmyHeo')
    
    if customer_name and rating_elem and review_text:
        review = {
            'customer_name': customer_name.text.strip(),
            'rating': rating_elem.text.strip()[0],
            'review_text': review_text.text.strip()
        }
        reviews.append(review)
    else:
        print('Some elements not found for a review.')

if __name__ == "__main__":
    print("Number of Reviews: ", len(reviews))
    for idx, review in enumerate(reviews, start=1):
        print(f"Review #{idx}:")
        print(f"Customer Name: {review['customer_name']}")
        print(f"Rating: {review['rating']}")
        print(f"Review: {review['review_text']}")
        print("\n"+ "="*50 + "\n")
else:
    print("No reviews were scraped.")
```

```python
import pandas as pd
df = pd.DataFrame(reviews)

# Print basic statistics
print("Average Rating:", df['rating'].astype(float).mean())
print("Distribution of Ratings:")
print(df['rating'].value_counts())
```

```python
df.to_csv('flipkart_scrape1_reviews.csv', index=False)
```
