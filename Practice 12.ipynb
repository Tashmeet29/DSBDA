```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('DSBDALExam DataSets/Adult/adult.csv', header=None)
column_names = ['age', 'workclass', 'fnlwgt', 'education', 'education-num', 'marital-status', 'occupation', 'relationship', 'race', 'sex', 'capital-gain', 'capital-loss', 'hours-per-week', 'native-country', 'loh50']
df.columns = column_names
df.index.name = 'Id'
```

```python
melted_data = pd.melt(df, id_vars=['age', 'workclass', 'education', 'race', 'sex'], value_vars=['native-country', 'loh50'], var_name='Information', value_name='Value')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))
sns.countplot(x='native-country', data=df)
plt.title('Counts of Individuals by Country')
plt.xlabel('Country')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
sns.pairplot(df, hue='sex', diag_kind='hist')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='sex', y='age', data=df)
plt.title('Age Distribution by Sex')
plt.xlabel('Sex')
plt.ylabel('Age')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(x='hours-per-week', hue='race', data=df, multiple='stack')
plt.title('Distribution of Work Hours per Week by Race')
plt.xlabel('Hours per Week')
plt.ylabel('Count')
plt.show()
```
