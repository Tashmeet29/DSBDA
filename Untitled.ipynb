```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import IsolationForest
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
df = pd.read_csv('DSBDALExam DataSets/Hepatitis/hepatitis.csv', header=None)
```

```python
df.columns = ['class', 'age', 'sex', 'steroid', 'antivirals', 'fatigue', 'malaise', 'anorexia', 'liver big', 'liver firm', 'spleen palpable', 'spiders', 'ascites', 'valices', 'bulirubin', 'alk phosphate', 'sgot', 'albomin', 'protime', 'histology']
```

```python
df.index.name = 'Id'
```

```python
df
```

```python
df.shape
```

# Data Cleaning

```python
df.replace('?', pd.NA, inplace=True)
```

```python
df.dropna(inplace=True)
```

```python
df = df.apply(pd.to_numeric, errors='coerce')
```

```python
df = df[(df >= 0).all(1)]
```

```python
df
```

```python
df.isna().sum()
```

```python
np.random.seed(42)
mask = np.random.rand(*df.shape) < 0.5
df[mask] = np.nan
```

```python
df.isna().sum()
```

```python
df.fillna(df.mode().iloc[0], inplace=True)
```

```python
df.isna().sum()
```

```python
df.isna().sum().sum()
```

```python
df
```

# Error Correcting / Outlier Removal

```python
iso = IsolationForest(contamination=0.1)
outliers = iso.fit_predict(df.drop('class', axis=1))
df = df[outliers == 1]
```

```python
df
```

# Data Transformation

```python
X = df.drop('class', axis=1)
y=df['class']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
lr = LogisticRegression()
lr.fit(X_train_scaled, y_train)
```

```python
y_pred_lr = lr.predict(X_test_scaled)
```

```python
accuracy_lr = accuracy_score(y_test, y_pred_lr)
```

```python
gnb = GaussianNB()
gnb.fit(X_train_scaled, y_train)
```

```python
y_pred_gnb = gnb.predict(X_test_scaled)
```

```python
accuracy_gnb = accuracy_score(y_test, y_pred_gnb)
```

```python
cm_lr = confusion_matrix(y_test, y_pred_lr)
cr_lr = classification_report(y_test, y_pred_lr, zero_division=1)
```

```python
cm_gnb = confusion_matrix(y_test, y_pred_gnb)
cr_gnb = classification_report(y_test, y_pred_gnb, zero_division=1)
```

```python
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
sns.heatmap(cm_lr, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.title('Logistic Regression Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.subplot(1,2,2)
sns.heatmap(cm_gnb, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.title('Gaussian NB Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.tight_layout()
plt.show()
```

```python
print('Logistic Regression Classification Report')
print(cr_lr)
```

```python
print('Gaussian NB Classification Report')
print(cr_gnb)
```

```python
print('Logistic Regression Accuracy Score')
print(accuracy_lr*100, '%')
```

```python
print('Gaussian NB Accuracy Score')
print(accuracy_gnb*100, '%')
```
