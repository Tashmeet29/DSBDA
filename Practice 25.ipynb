```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df.hist(figsize=(10,12))
plt.tight_layout()
plt.show()
```

```python
sns.pairplot(df, hue='sex', diag_kind='hist')
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(df['steroid'], df['fatigue'], alpha=0.5)
plt.title('Steroid vs Fatigue')
plt.xlabel('Steroid')
plt.ylabel('Fatigue')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(df['sgot'], df['histology'], alpha=0.5, edgecolors='green')
plt.title('Sgot vs Histology')
plt.xlabel('Sgot')
plt.ylabel('Histology')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.countplot(x='age', data=df)
plt.title('Counts of Individuals by Age')
plt.xlabel('Age')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```
